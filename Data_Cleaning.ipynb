```python
import kagglehub
import pandas as pd
```

```python
df = pd.read_excel(r'/content/Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns='Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('./_')
df['Last_Name']
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[:3] +'-'+ x[3:6] +'-'+ x[6:])
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
df[['street_address','state','zipcode']] = df['Address'].str.split(',',n=2, expand=True)
df
```

```python
df['Paying Customer'].value_counts()
```

```python
df['Do_Not_Contact'].value_counts()
```

```python
df['Paying Customer']= df['Paying Customer'].str.replace('yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.strip()
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df = df.fillna('')
df
```

```python
df = df[~(df['Do_Not_Contact']=='Y')]
df = df[~(df['Phone_Number']=='')]
df = df.reset_index(drop=True)
df
```
